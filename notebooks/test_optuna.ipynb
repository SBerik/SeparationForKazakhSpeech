```python
import optuna

def objective(trial):
    x = trial.suggest_float("x", -10, 10)  # Search space for x
    return (x - 2) ** 2  # Example function to minimize

study = optuna.create_study(direction="minimize")  # Minimize the function
study.optimize(objective, n_trials=100)  # Run 100 trials

print("Best hyperparameters:", study.best_params)
print("Best value:", study.best_value)
```
